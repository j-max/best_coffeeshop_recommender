```python
import pandas as pd
import numpy as np
import pickle

import matplotlib.pyplot as plt
```

```python
with open('../models/dataframes/coffee_per_zip.p', 'rb') as read_file:
    df = pickle.load(read_file)
```

```python
df.head()
```

```python
df.set_index('zipcodes', inplace=True)
```

```python
df.mean_zip_income = df.mean_zip_income.str.replace(',', '')
df.fillna(np.nan, inplace=True)
df.mean_zip_income = df.mean_zip_income.astype(float)
mean_mean_zip_inc = df['mean_zip_income'].mean()
df.mean_zip_income.fillna(mean_mean_zip_inc, inplace=True)
```

```python
df.people_per = df.people_per.str.replace(',', '')
df.fillna(np.nan, inplace=True)
df.people_per = df.people_per.astype(float)
mean_people_per = df['people_per'].mean()
df.people_per.fillna(mean_people_per, inplace=True)
```

## Kmeans

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
num_clusters =10
```

```python
km = KMeans(n_clusters=num_clusters, random_state=41)
```

```python
X
```

```python
km.fit(X)
```

```python
km.cluster_centers_
```

```python
coffee_shop_count_centers = [center[0] for center in km.cluster_centers_]
clusters = list(range(1, num_clusters+1))
len(coffee_shop_count_centers)
plt.title('Coffee Shop Cluster Centers')
plt.bar(clusters,coffee_shop_count_centers)
```

```python
rides_per_el_centers = [center[1] for center in km.cluster_centers_]
clusters = list(range(1, num_clusters+1))
len(rides_per_el_centers)
plt.title('Rides per day closest EL')
plt.bar(clusters,rides_per_el_centers)
```

```python
coord_dist_to_el_center = [center[2] for center in km.cluster_centers_]
clusters = list(range(1, num_clusters+1))
len(coord_dist_to_el_center)
plt.title('Coord Dist to El')
plt.bar(clusters,coord_dist_to_el_center)
```

```python
busi_count_center = [center[3] for center in km.cluster_centers_]
clusters = list(range(1, num_clusters+1))
len(busi_count_center)
plt.title('Business Count Closest to El')
plt.bar(clusters,busi_count_center)
```

```python
mean_zip_income_center = [center[4] for center in km.cluster_centers_]
clusters = list(range(1, num_clusters+1))
len(mean_zip_income_center)
plt.title('Mean Zips')
plt.bar(clusters,mean_zip_income_center)
```

```python
people_per = [center[5] for center in km.cluster_centers_]
clusters = list(range(1, num_clusters+1))
len(people_per)
plt.title('People per Center')
plt.bar(clusters,people_per)
```

```python
zips = df.index
for zipcode, cluster in zip(zips, km.labels_):
    print(zipcode, cluster)
```

```python
for _ in range(1,10):
    for zipcode, cluster in zip(zips,km.labels_):
        if _ == cluster:
            print(zipcode, cluster)
```
