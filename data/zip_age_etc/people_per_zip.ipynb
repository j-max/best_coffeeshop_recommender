```python
from urllib import request
from bs4 import BeautifulSoup
import re
import pickle
import time
```

```python
!ls ..
```

```python
with open('../zip_codes/zip_lat_long.pk', 'rb') as read_file:
    zip_codes = pickle.load(read_file)
```

```python
zip_codes = list(zip_codes.keys())
len(zip_codes)
```

```python
people_per = {}
for zip_code in zip_codes:
    try:
        url = 'https://censusreporter.org/profiles/86000US%s-%s' %(zip_code,zip_code)
        req = request.Request(url)
        with urllib.request.urlopen(req) as response:
           the_page = response.read()
        soup = BeautifulSoup(the_page,'html.parser') 
        age = soup.find_all('span', attrs = {'class':'secondary'})
        people_pattern = re.compile('[\d]+[,][\d]+[.][\d]*')
        people=people_pattern.search(age[1].text)
        people_per[zip_code] = people[0]
        time.sleep(.5)
        print(people_per)
    except:
        pass
```

```python
with open('people_per_zip.pk', 'wb') as write_file:
    pickle.dump(people_per, write_file)
```
