```python
blocks_df =  pd.read_csv('CensusBlockTIGER2010.csv')
blocks_df
```

```python
print(blocks_df.iloc[0]['the_geom'])
```

```python
import re
```

```python
one_block = blocks_df.iloc[1]['the_geom'].split(" ")
lat_pat = re.compile('[-][\d]+[.][\d]+')
lats = lat_pat.findall(str(one_block))
long_pat = re.compile('[^-\d][\d]+[.][\d]+')
longs = long_pat.findall(str(one_block))
```

```python
float_lats = [float(x) for x in lats]
avg_lat = sum(float_lats)/len(float_lats)
float_longs = [float(x[1:]) for x in longs]
avg_long = sum(float_longs)/len(float_longs)
```
