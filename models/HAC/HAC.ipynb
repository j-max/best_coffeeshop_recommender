```python
import pandas as pd
import numpy as np
import pickle
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
```

```python
with open('../dataframes/df_with_libs.p','rb') as read_file:
    df = pickle.load(read_file)
    
#This census block's Total Pop was in error (Over 9000)
df = df[df['CENSUS BLOCK']!=8435001023]
```

```python
X = df.drop(columns=['CENSUS BLOCK', 'closest_el', 'daily_rides_closest_el','block_lat', 'block_lon'])
```

## Linkage Single

```python
clustering = AgglomerativeClustering(linkage='single', n_clusters=7)
clustering.fit(X)
```

```python
clustering.labels_
```

```python
pca = PCA(n_components = 2)
pca_result = pca.fit_transform(X)
pca_one = pca_result[:,0]
pca_two = pca_result[:,1]
```

```python
labels = clustering.labels_
labels_, counts = np.unique(labels, return_counts=True)
labels_
plt.figure(figsize=[10,10])
# legend = np.linspace(-1,len())

plt.scatter(pca_one, pca_two, c=clustering.labels_, s=5)
```

```python
clustering = AgglomerativeClustering(linkage='complete', n_clusters=3)
clustering.fit(X)
```

```python
X = df.drop(columns=['CENSUS BLOCK', 'closest_el', 'daily_rides_closest_el','block_lat', 'block_lon'])
```

```python
hac_clust_7 = AgglomerativeClustering(linkage='complete', n_clusters=7)
hac_clust_7.fit(X)
```

```python
X = df.drop(columns=['CENSUS BLOCK', 'closest_el', 'daily_rides_closest_el','block_lat', 'block_lon'])
X = StandardScaler().fit_transform(X)
pca = PCA(n_components = 3)
pca_result = pca.fit_transform(X)
pca_one = pca_result[:,0]
pca_two = pca_result[:,1]
df_hac7 = df
df_hac7['hac'] = hac_clust_7.labels_
df_hac7['pca_one'] = pca_one
df_hac7['pca_two'] = pca_two

labels = np.unique(hac_clust_7.labels_)

for dblabel in np.unique(labels):
    df_for_plot = df_hac7[df_hac7.hac == dblabel]
    df_pca_one = df_for_plot.pca_one
    df_pca_two = df_for_plot.pca_two
    plt.scatter(df_pca_one, df_pca_two, s=5, label=dblabel)
plt.title("HAC")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
# plt.legend()

# labels = hac_clust_7.labels_
# labels_, counts = np.unique(labels, return_counts=True)
# labels_
# plt.figure(figsize=[10,10])
# # legend = np.linspace(-1,len())

# plt.scatter(pca_one, pca_two, c=hac_clust_7.labels_, s=5)
# plt.xlabel('Principal Component 1')
# plt.ylabel('Principal Component 2')
# plt.title('HAC - 7 Clusters')
plt.savefig('../../visualization/HAC_clusters/HAC_7clust.svg')
```
